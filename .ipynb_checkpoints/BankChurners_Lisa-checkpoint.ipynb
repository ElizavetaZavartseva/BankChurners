### Perform Train Test Split

```python
features = df.drop(columns = ["clientnum", "attrition_flag"])
target = df["attrition_flag"]
```

```python
features.head(5)
```

```python
target.head(5)
```

```python
target.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

# Bagging and Pasting

### For original dataset 

```python
bagging_class = BaggingClassifier(DecisionTreeClassifier(max_depth=20),
                               n_estimators=100,
                               max_samples = 1000)

bagging_class.fit(X_train, y_train)


pred = bagging_class.predict(X_test)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", bagging_class.score(X_test, y_test))
```

### For norm dataset 

```python
bagging_class = BaggingClassifier(DecisionTreeClassifier(max_depth=20),
                               n_estimators=100,
                               max_samples = 1000)

bagging_class.fit(X_train_norm, y_train)

pred = bagging_class.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", bagging_class.score(X_test_norm, y_test))
```

### For reduced dataset 

```python
bagging_class = BaggingClassifier(DecisionTreeClassifier(max_depth=20),
                               n_estimators=100,
                               max_samples = 1000)

bagging_class.fit(X_train_reduced, y_train)

pred = bagging_class.predict(X_test_reduced)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", bagging_class.score(X_test_reduced, y_test))
```

### For norm dataset 

```python
forest_norm = RandomForestClassifier(n_estimators=100,
                             max_depth=20)

forest_norm.fit(X_train_norm, y_train)

pred = forest_norm.predict(X_test_norm)

print("MAE", mean_absolute_error(pred, y_test))
print("RMSE", mean_squared_error(pred, y_test, squared=False))
print("R2 score", forest_norm.score(X_test_norm, y_test))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming X_train_norm is a DataFrame or has named columns
feature_names = X_train_norm.columns

# Get feature importances
importances = forest_norm.feature_importances_
indices = np.argsort(importances)[::-1]

# Arrange the feature names in the order of their importance
sorted_feature_names = [feature_names[i] for i in indices]

# Specify colors (e.g., using 'skyblue' for all bars)
colors = ['skyblue'] * len(importances)

# Plot the feature importances with specified colors
plt.figure(figsize=(10, 8))
plt.title("Feature Importances")
colors = [cmap(i / len(importances)) for i in range(len(importances))]

plt.bar(range(X_train_norm.shape[1]), importances[indices], align="center", color=colors)
plt.xticks(range(X_train_norm.shape[1]), sorted_feature_names, rotation=90)
plt.xlabel('Feature Names')
plt.ylabel('Importance')
plt.tight_layout()
plt.show()
```

```python
results = model.cv_results_

# Create a DataFrame to display the results
results_df = pd.DataFrame({
    'params': results['params'],
    'mean_test_score': results['mean_test_score'],
    'std_test_score': results['std_test_score'],
    'rank_test_score': results['rank_test_score']
})

# Print the DataFrame
print(results_df)
```

```python
import seaborn as sns

# Assuming 'max_depth' and 'n_estimators' are the parameters we want to visualize
# Extracting the values for heatmap plotting
results_df['max_depth'] = results_df['params'].apply(lambda x: x['max_depth'])
results_df['n_estimators'] = results_df['params'].apply(lambda x: x['n_estimators'])

# Pivot table for heatmap
pivot_table = results_df.pivot(index='max_depth', columns='n_estimators', values='mean_test_score')

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, fmt=".5f", cmap='viridis')
plt.title('Mean Test Score Heatmap')
plt.xlabel('Number of Estimators')
plt.ylabel('Max Depth')
plt.show()
```
